```python
#3
def powp(n):
    if n==1:
        return 1
    elif n>0:
        return n**2+powp(n-1)
    else:
        print("非法输入")
```

```python
powp(10)
```

```python
powp(100)
```

```python
powp(-2)
```

```python
def reverse_(s):
    if len(s)==1:
        return s
    else:
        l=list(s)
        n=l.pop()
        return n+reverse_("".join(l))
```

```python
#4
def rise():
    total=0
    def inner(m,reset=0):
        nonlocal total
        total+=m
        if reset==True:
            total=0
        return total
    return inner
```

```python
x=rise()
```

```python
x(100)
```

```python
x(1000)
```

```python
x(0,reset=1)
```

```python
#5
def pei(n):
    i=1
    y=[]
    back1,back2=0,1
    while i<n: 
        y.append(back1)
        back1,back2=back2,back1+back2
        i+=1
    z=filter(lambda x: x%2==0,y)
    return list(z)
```

```python
pei(50)
```

```python
def leix():
    def inner(n):
        if n==1:
            return 1
        elif n>=2:
            return n*inner(n-1)
        else:
            print("非法输入")
    return inner
```

```python
q=leix()
```

```python
q(100)
```

```python
q(10)
```

```python
def bigger(*args):
    avg=sum(args)/len(args)
    x=filter(lambda x: x>avg,args)
    return list(x)
```

```python
bigger(11,21,34,21,12,43,24,17)
```
