```python
#6.
def del_idx_str(s,idx):
    if idx>=len(s):
        print("请输入字符串的索引")
    else:
        l=s[:idx]+s[idx+1:len(s)]
        return l
```

```python
del_idx_str("hello",3)
```

```python
#四 7
def inner(s):
    if len(s)==1:
        return s
    else:
        l=list(s)
        m=l.pop(len(l)-1)
        return m+inner("".join(l))
    
```

```python
inner("hello")
```

```python
#8
def pow_(n):
    if n==1:
        return 1
    elif n%2==0:
        n=n-1
        return n*n+pow_(n-2)
    else:
        return n*n+pow_(n-2)
```

```python
pow_(10)
```

```python
#10
def zhishu():
    for n in range(1,101):
        if n<2:
            continue
        factors={x:n//x for x in range(1,n+1) if n%x==0}
        if len(factors)==2:
            yield n
```

```python
y=zhishu()
```

```python
next(y)
```

```python
next(y)
```

```python
def outter():
    n=0
    def count():
        global n
        n+=1
        return n
    def reset():
        global n
        n=0
        print("清零")
    return count,reset
```
