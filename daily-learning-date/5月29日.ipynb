```python
#构造函数 __init__()
```

```python
class G:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def add(self):
        return self.x+self.y
    def mul(self):
        return self.x*self.y
```

```python
g=G(2,3)
```

```python
g.add()
```

```python
g.mul()
```

```python
g.__dict__
```

```python
#重写
```

```python
class H(G):
    def __init__(self,x,y,z):
        G.__init__(self,x,y,)
        self.z=z
    def add(self):
        return G.add(self)+self.z
    def mul(self):
        return G.mul(self)*self.z
```

```python
h=H(2,3,4)
```

```python
h.add()
```

```python
h.mul()
```

```python
#钻石继承
```

```python
class I:
    def __init__(self):
        print("我是I")
class J1(I):
    def __init__(self):
        I.__init__(self)
        print("我是J1")
class  J2(I):
    def __init__(self):
        I.__init__(self)
        print("我是J2")
class K(J1,J2):
    def __init__(self):
        J1.__init__(self)
        J2.__init__(self)
        print("我是K")
```

```python
k=K()
```

```python
#super()
```

```python

class J1(I):
    def __init__(self):
        super().__init__()
        print("我是J1")
class  J2(I):
    def __init__(self):
        super().__init__()
        print("我是J2")
class K(J1,J2):
    def __init__(self):
        super().__init__()
        print("我是K")
```
