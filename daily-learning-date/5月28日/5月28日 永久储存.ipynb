```python
import pickle
x,y,z=1,2,3
s="python"
l=["hello",114,1.43]
d={"one":1,"two":2}
with open("data.pkl","wb")as f:
    pickle.dump(x,f)
    pickle.dump(y,f)
    pickle.dump(z,f)
    pickle.dump(l,f)
    pickle.dump(s,f)
    pickle.dump(d,f)
```

```python
with open("data.pkl","rb")as f:
    x=pickle.load(f)
    y=pickle.load(f)
    z=pickle.load(f)
    d=pickle.load(f)
    l=pickle.load(f)
    s=pickle.load(f)
print(x,y,z,s,l,d,sep="\n")
```
