```python
#一。1
def calc_avg(nums):
    """
    功能：接受一个数字列表，返回列表的平均值
    参数：
    -nums 输入的的数字列表
    -len(nums) 列表元素数量
    - return 输出的平均值，当nums为空列表是输出None
    示例：
    输入：calc_avg([1,2,3])
    输出：2
    """
    if len(nums)==0:
        return None
        print("不能输入空列表")#print不能放在return None后面
    else:
        return sum(nums)/len(nums)
    
```

```python
#二.3
def hanoi(n,A,B,C):
    if n==1:
        print(f"盘子{n}：柱子{A}->柱子{C}")
    else:
        hanoi(n-1,A,C,B)
        print(f"盘子{n}：柱子{A}->柱子{C}")
        hanoi(n-1,B,A,C)
```

```python
hanoi(3,'x','y','z')
```
