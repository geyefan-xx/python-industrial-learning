```python
#1
import math
def math_calc(calc_type,decimal=2,*data):
    if all(isinstance(x,(int,float)) for x in data):
        print("请至少传入一个数字")
        return None
    if calc_type=="sum":
        r=sum(data)
    elif calc_type=="avg":
        r=sum(data)/len(data)
    elif calc_type=="prod":
        r=math.prod(data)
    else:
        print("mode参数只能是sum、avg、prod")
        return None
    print("计算结果为：{:.{}f}".format(r,decimal))
```

```python
math_calc("prod",3,12,4,6,2,7)
```

```python
#2
def shop_create(shop_name):
    t=0
    def set_discount(dis):
        def inner(price):
            nonlocal t
            t=dis*price+t
            return t
            print(shop_name,f"当前折后累计营收：{t:.2f}元")
        return inner
    return set_discount
```

```python
xiaoshoue=shop_create("和平饭店")
```

```python
xiaoshoue(dis=0.8)(price=100)
```

```python
#3
import time
def myfunc(role):
    def midfunc(func):
        def inner(*args,**kwargs):
            if role=="admin":
                start=time.time()
                func(*args,**kwargs)
                end=time.time()
                print(func.__name__,"共耗时：",start-end,"s")
            elif role=="user":
                func(*args,**kwargs)
            else:
                print("权限不足，禁止调用")
                return None
        return inner
    return midfunc
```

```python
@myfunc(role="admin")
def add(*args):
    Sum=sum(args)
    print(Sum)
```

```python
add(1,3,5,7,9)
```

```python
#4
def deco1(func):
    def inner():
        print("数据初始化完成")
        func()
    return inner
def deco2(func):
    def inner():
        print("-------------")
        func()
        print("-------------")
    return inner
def deco3(func):
    def inner():
        func()
        print("任务执行完毕")
    return inner
```

```python
@deco3
@deco2
@deco1
def myfunc():
    print("函数执行中")
```

```python
myfunc()
```
