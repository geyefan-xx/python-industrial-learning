```python
#1
def calc_all(*nums,**opts):
    mode=opts.get("mode","avg")
    precision=opts.get("precision",2)
    if mode=="sum":
        result=sum(nums)
    if mode=="avg":
        result=sum(nums)/len(nums)
    print("计算结果为：{:.{}f}".format(result,precision))
```

```python
calc_all(1,2,3,1,4,mode="avg",precision=3)
```

```python
#2
def base_salary(base):
    def inner(merit_pay):
        salary=base+merit_pay
        return salary
    return inner
```

```python
manager=base_salary(8000)
print(manager(2500))
```

```python
worker=base_salary(5000)
print(worker(1200))
```

```python
#3
def funcA(func):
    def inner(*args,**kwargs):
        print("【开始执行】",func.__name__)
        result=func(*args,**kwargs)
        print("【执行完毕】",func.__name__)
        return result
    return inner
```

```python
@funcA
def pow(x):
    r=x**x
    return r
```

```python
pow(5)
```

```python
#4
def price():
    total=0
    def inner(x):
        nonlocal total
        total=total+x
        return total
    return inner
```

```python
total_price=price()
```

```python
total_price(125)
```

```python
total_price(648)
```

```python
total_price(520)
```

```python
#5
stu_score={"小明":{"语文":78,"数学":92},"小红":{"语文":85,"数学":76}}
```

```python
adv_stu={s:sum(v.values())/len(v) for s,v in stu_score.items()}
```

```python
adv_stu
```

```python
exelent_stu={s:v for s,v in adv_stu.items() if v>80}
```

```python
exelent_stu
```

```python
for s,v in adv_stu.items():
    print(f"{s}同学的平均分为：{v:.2f}")
```
